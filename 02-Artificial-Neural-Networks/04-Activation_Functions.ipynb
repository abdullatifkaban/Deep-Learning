# Yapay Sinir Ağlarında Aktivasyon Fonksiyonları

Bu notebook, **Sigmoid**, **ReLU**, **Tanh** ve **Softmax** gibi yaygın aktivasyon fonksiyonlarının nasıl çalıştığını göstermektedir.
Her fonksiyonun matematiksel tanımı, grafiği ve Python ile uygulanışını içermektedir.



```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid fonksiyonu
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU fonksiyonu
def relu(x):
    return np.maximum(0, x)

# Tanh fonksiyonu
def tanh(x):
    return np.tanh(x)

# Softmax fonksiyonu
def softmax(x):
    e_x = np.exp(x - np.max(x))  # Sayısal kararlılığı artırmak için max çıkarılır
    return e_x / e_x.sum(axis=0)

# X değer aralığı oluştur
x = np.linspace(-6, 6, 100)
```

## Sigmoid Fonksiyonu

Sigmoid fonksiyonu, çıkış değerini (0,1) aralığına sıkıştıran sürekli bir fonksiyondur. Genellikle olasılık tahmini gerektiren modellerde kullanılır.

Matematiksel gösterimi:

$$ f(z) = \frac{1}{1+e^{-z}} $$

Sigmoid fonksiyonunun grafiği aşağıda verilmiştir:


```python
plt.figure(figsize=(6,4))
plt.plot(x, sigmoid(x), label="Sigmoid", color='blue')
plt.title("Sigmoid Fonksiyonu")
plt.xlabel("Giriş Değeri (z)")
plt.ylabel("Çıkış Değeri f(z)")
plt.grid()
plt.legend()
plt.show()
```

## ReLU (Rectified Linear Unit) Fonksiyonu

ReLU fonksiyonu, negatif girişleri sıfıra ayarlar ve pozitif girişleri olduğu gibi bırakır.

Matematiksel gösterimi:

$$ f(z) = \max(0, z) $$

ReLU fonksiyonunun grafiği aşağıda verilmiştir:


```python
plt.figure(figsize=(6,4))
plt.plot(x, relu(x), label="ReLU", color='red')
plt.title("ReLU Fonksiyonu")
plt.xlabel("Giriş Değeri (z)")
plt.ylabel("Çıkış Değeri f(z)")
plt.grid()
plt.legend()
plt.show()
```

## Tanh Fonksiyonu

Tanh fonksiyonu, giriş değerini -1 ile 1 arasında sınırlar. Sigmoid fonksiyonuna kıyasla sıfır merkezlidir.

Matematiksel gösterimi:

$$ f(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}} $$

Tanh fonksiyonunun grafiği aşağıda verilmiştir:


```python
plt.figure(figsize=(6,4))
plt.plot(x, tanh(x), label="Tanh", color='green')
plt.title("Tanh Fonksiyonu")
plt.xlabel("Giriş Değeri (z)")
plt.ylabel("Çıkış Değeri f(z)")
plt.grid()
plt.legend()
plt.show()
```

## Softmax Fonksiyonu

Softmax fonksiyonu, çok sınıflı sınıflandırma problemlerinde kullanılır ve her sınıfa ait olasılığı hesaplar.

Matematiksel gösterimi:

$$ f(z_i) = \frac{e^{z_i}}{\sum_{j} e^{z_j}} $$

Softmax fonksiyonunun grafiği aşağıda verilmiştir:


```python
softmax_outputs = softmax(np.vstack([x, x + 2, x - 2]))  # Üç farklı giriş vektörü için

plt.figure(figsize=(6,4))
plt.plot(x, softmax_outputs.T, label=["Sınıf 1", "Sınıf 2", "Sınıf 3"])
plt.title("Softmax Fonksiyonu")
plt.xlabel("Giriş Değeri (z)")
plt.ylabel("Olasılık Dağılımı")
plt.legend()
plt.grid()
plt.show()
```

## Sonuç

Bu çalışmada, yapay sinir ağlarında kullanılan temel aktivasyon fonksiyonlarını inceledik.  
Her fonksiyonun avantajları ve dezavantajları şunlardır:

- **Sigmoid**: Küçük değerlerde iyi çalışır, ancak gradyan kaybolma problemi yaşayabilir.
- **ReLU**: Hesaplama açısından verimli ve genellikle derin ağlarda tercih edilir, ancak ölü nöron problemi olabilir.
- **Tanh**: Sigmoid’e benzer, ancak sıfır merkezlidir ve daha iyi öğrenme performansı sunabilir.
- **Softmax**: Çok sınıflı sınıflandırma problemlerinde olasılık dağılımı oluşturmak için kullanılır.

Aktivasyon fonksiyonlarının doğru seçimi, modelin başarısı için kritik öneme sahiptir.
